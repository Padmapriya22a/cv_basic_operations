```python
import cv2
```

```python
path='tom&jerry.jpg'
```

```python
tj=cv2.imread(path,1)
```

```python
cv2.imshow('image',tj)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(tj.shape)
print(tj.size)
print(tj.dtype)
```

```python
pix=tj[50,50]
print(pix)
```

```python
b,g,r=cv2.split(tj) # split into BGR channels 
print(b,g,r)
bl=tj[:,:,0] # split using numpy into BGR channels B-0,G-1,R-2
print('Blue')
print(bl)
```

```python
img=cv2.merge((b,g,r)) # merge the BGR channels
cv2.imshow('merge',img) 
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
mo=cv2.imread(path,1)
mo[:,:,0]=0
cv2.imshow('Modified_color',mo)
cv2.imwrite('Modified_color.png',mo)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('CO',img)
cv2.imwrite('COLOR.png',img)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
for n in range (1,50):
    for m in range (1,50):
        for i in range (0,3):
            print(tj.item(m,n,i))
            tj.itemset((m,n,i),255)
cv2.imshow('Modified_pixel',tj)
cv2.imwrite('Modified_pixel.png',mo)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread(path,1)
img=cv2.copyMakeBorder(img,5,5,5,5,cv2.BORDER_REPLICATE,value=[255,255,255])
cv2.imshow('im',img)
cv2.imwrite('border.png',img)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import numpy
import os

randomByteArray= bytearray(os.urandom(120000))
flatNumpyArray=numpy.array(randomByteArray)
gray=flatNumpyArray.reshape(300,400)
cv2.imshow('img_gen',gray)
cv2.waitKey(0)
cv2.imwrite('img_gen.png',gray)
bgr=flatNumpyArray.reshape(200,200,3)
cv2.imshow('img_gen1',bgr)
cv2.imwrite('img_gen1.png',bgr)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

# combine 2 images

```python
gy=cv2.imread("D:/tech/Untitled Folder/galaxy.jpg",1)
ny=cv2.imread("D:/tech/Untitled Folder/noddy.jpg",1)
```

```python
print(gy.shape)
print(ny.shape)
```

```python
gy=cv2.resize(gy,(256,256))
ny=cv2.resize(ny,(256,256))
cv2.imshow('image',gy)
cv2.waitKey(0)
cv2.imshow('image1',ny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Add

```python
add=cv2.add(gy,ny);
cv2.imshow('image',add)
cv2.waitKey(0)
cv2.imwrite('combined.png',add)
cv2.destroyAllWindows()
```

```python
add_w=cv2.addWeighted(gy,0.7,ny,0.3,0);
cv2.imshow('image',add_w)
cv2.waitKey(0)
cv2.imwrite('combined_w.png',add_w)
cv2.destroyAllWindows()
```

### Bitwise

```python
bitand=cv2.bitwise_and(gy,ny);
cv2.imshow('image',bitand)
cv2.waitKey(0)
cv2.imwrite('combined_BA.png',bitand)
cv2.destroyAllWindows()
```

```python
bitor=cv2.bitwise_or(gy,ny);
cv2.imshow('image',bitor)
cv2.waitKey(0)
cv2.imwrite('combined_Bo.png',bitor)
cv2.destroyAllWindows()
```

```python
bitxor=cv2.bitwise_xor(gy,ny);
cv2.imshow('image',bitxor)
cv2.waitKey(0)
cv2.imwrite('combined_Bxo.png',bitxor)
cv2.destroyAllWindows()
```
